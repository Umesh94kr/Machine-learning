```python
import numpy as np
import openpyxl
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def feature_changing(x_train):
    x_train['internet'] = x_train['internet'].map({'no' : 0, 'yes' : 1})
    x_train['sex'] = x_train['sex'].map({'F' : 1, 'M' : 2})
```

```python
train_data = pd.read_excel('Training data.xlsx')
feature_changing(train_data)
test_data = pd.read_excel('Test data.xlsx')
feature_changing(test_data)

x_train = np.array(train_data.iloc[:,0:8])
y_train = np.array(train_data.iloc[:,8]).reshape(-1,1)
```

```python
def z_score(x_train):

  x_mean = np.mean(x_train, axis=0)
  x_std = np.std(x_train, axis=0)
  x_train = (x_train - x_mean)/x_std

  return x_train,x_std,x_mean
```

```python
def model(x_train, w, b):
  return (np.dot(x_train, w) + b)
```

```python
def cost(x_train,y_train,w,b):
  y_pred = model(x_train, w, b)
  m = len(x_train)
  loss = 0
  for i in range(m):
    loss += (y_pred[i] - y_train[i])**2
  loss = loss/(2*m)
  return loss
```

```python
def gradient_descent(x_train,y_train,w,b):
  y_pred = model(x_train, w, b)
  total_cost = cost(x_train, y_train, w, b)
  print(total_cost)
  dw = np.dot(x_train.T, (y_pred - y_train))
  db = np.sum(y_pred - y_train)

  dw /= len(x_train)
  db /= len(x_train)

  alpha = 0.01
  
  w -= alpha * dw
  b -= alpha * db


  return w,b
```

```python
x_train = x_train.astype(np.float64)
x_train,x_std,x_mean = z_score(x_train)

np.random.seed(2147483647)
w = np.random.randn(x_train.shape[1],1)
b = np.random.randn(1)

old_cost = 0

while abs(old_cost - cost(x_train,y_train,w,b))>0.00001:
  old_cost = cost(x_train,y_train,w,b)
  w,b = gradient_descent(x_train,y_train,w,b)


x_predict = test_data.iloc[:,:8].to_numpy()
x_predict = (x_predict - x_mean)/x_std
ans = test_data.iloc[:,8].to_numpy()

y_predict = np.dot(x_predict,w) + b

accuracy = 0
for dim in range(len(ans)):
  if abs(y_predict[dim]-ans[dim])<0.5: # do not change the tolerance as you'll be checked on +- 0.5 error only
    accuracy += 1
accuracy = round(accuracy*100/200.0,2)
ok = 'Congratulations' if accuracy>95 else 'Optimization required'
print(f"{ok}, your accuracy is {accuracy}%")
```
